#Finding e-mails using regular expressions






```python
from selenium import webdriver
import re

driver=webdriver.Chrome()
driver.get('http://www.airindia.in/customer-support.htm')

doc=driver.page_source
email=re.findall(r'[\w\.-]+@[\w\.-]*',doc)
phones=re.findall(r'[\d]{10}',doc)

#print(len(email))
for i,j in zip(email,phones):
    print(i,j)
```
